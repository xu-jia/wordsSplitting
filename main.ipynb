```python
import json
# function for generating a list of tuples that contains the index of words
def get_idx(json_file):
    with open(json_file) as f:
        data = json.load(f)   
    res = []
    for sent in data:
        start = 0
        tmp = []
        for word in sent:
            tmp.append((start,start+len(word)-1))
            start = start+len(word)
        res.append(tmp)
    return res,data
```

```python
# get all index of resulting files, as well as the reference words splitting
json_file = "fdata/ref.json"
ref,ref_data = get_idx(json_file)

json_file = "fdata/fmm.json"
res_fmm,fmm_data = get_idx(json_file)

json_file = "fdata/mm.json"
res_mm,mm_data = get_idx(json_file)

json_file = "fdata/jieba.json"
res_jieba,jieba_data = get_idx(json_file)

json_file = "fdata/sougou.json"
res_sougou,sougou_data = get_idx(json_file)

json_file = "fdata/ltp.json"
res_ltp,ltp_data = get_idx(json_file)

json_file = "fdata/lac.json"
res_lac,lac_data = get_idx(json_file)
```

#### Now, we can compare the results obtained with different methods and the reference words splitting

```python
def len_verify(f1,f2):
    dif = []
    for i in range(len(f1)):
        if f1[i][-1][-1]!=f2[i][-1][-1]:
            dif.append(i)
    return dif,len(dif)
```

```python
# 比较句子分词的字符长度
print("Number of sentences in which the length differs from reference")
print("fmm:",len_verify(ref,res_fmm)[1])
print("mm:",len_verify(ref,res_mm)[1])
print("lac:",len_verify(ref,res_lac)[1])
print("ltp:",len_verify(ref,res_ltp)[1])
print("jieba:",len_verify(ref,res_jieba)[1])
print("sougou:",len_verify(ref,res_sougou)[1])
```

- punctuation not considered in sougou.
- The other methods do not take words in alphabet into account, as compares in the following cells:

```python
def display_diff(diff,c_res,c_data):
    """
    diff: index of sentences that differs from reference
    c_res: resulting index of words splitting (res_fmm, res_mm, res_jieba, res_sougou, res_ltp, res_lac)
    c_data: words splitting (fmm_data, mm_data, jieba_data, sougou_data, ltp_data, lac_data)
    """
    for i in diff:
        print(f"##sent_id = {i+1}".format(i))
        print(100*"-")
        print([ref_data[i][n]+str(ref[i][n]) for n in range(len(ref_data[i]))])
        print([c_data[i][n]+str(c_res[i][n]) for n in range(len(c_res[i]))])
        print(100*"-")
```
